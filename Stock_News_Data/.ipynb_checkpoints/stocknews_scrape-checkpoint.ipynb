```python
#get stock price information.
#an average of 1 day gain/loss, 5 day, 1 month 
import yfinance as yf
from datetime import datetime
from datetime import timedelta
# Get the data for the stock AAPL



def loss_gain(ticker, date, end = 0, less = False):
    
    close_list = []
    if not less: 
        for n in range(1,30-2,1): 
            done = date + timedelta(days=n)
            day = yf.download(ticker,date, done)
            first = day.iloc[0].Close
            last = day.iloc[-1].Close
            close_list += [(last - first)/first]
    else:
        for n in range(1,end-2,1): 
            done = date + timedelta(days=n) 
            day = yf.download(ticker,date, done)
            first = day.iloc[0].Close
            last = day.iloc[-1].Close
            close_list += [(last - first)/first]
    max_pct_change = max(close_list)
    future_day = close_list.index(max_pct_change)
    return [[future_day, 100 * max_pct_change]]
#getting ticker name   

change = []
for date in df.index: 
    ticker = df.loc[date].ticker

    #getting date-> reformatting 
    date = datetime.strptime(date, '%B %d, %Y').strftime('%Y-%m-%d')
    date = datetime.strptime(date, '%Y-%m-%d' )
    end = date - datetime.now()
    end = abs(int(str(end).split(',')[0].replace('days','')))
    #getting highest percent change with future days of change 

    if end > 30: 
        change += loss_gain(ticker, date)
    else: 
        change += loss_gain(ticker, date, end, less=True)
    
    
```
